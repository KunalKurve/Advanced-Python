Regular expressions are used for pattern matching

to design regular expression we use the following symbols

Use https://regex101.com/ for practice

. : it matches with 1 character

\w : it matches with any one word character [ a-z, A-Z0-9_]

\W : it matches with any one word character other than alphabet, digits, underscore[^a-z,A-Z0-9_]

\D : any non digit character

\b : any one boundary character

\B: any one boundary character

* : 0 or more occuences

+ : 1 or more occurences

{n} exactly n occurences

{m,n} minimum m occurences, maximum n occurences

{m,} minimum m occurences

^ : it matches the pattern at the beginning of the string

$ : matches the pattern at the end of the string


We can validate input data from User via REGEX

LIKE FOR MOBILE NUMBER : ^\d{10}$

```python
s = "Something is there somewhere"

import re

m = re.search("s.*e",s,re.I) #search finds out only the first occurrance
if m != None: #m != None means that the pattern exists in the string
  print("Found")
  print(m.group()) #returns the group(strng of characters) where the pattern was found
  print(m.span()) #span returns the position where the pattern group was found, returns a tuple of range
else:
  print("not found")

m = re.search("t.*?e",s,re.I) #search finds out only the first occurrance
if m != None: #m != None means that the pattern exists in the string
  print("Found")
  print(m.group()) #returns the group where the pattern was found
  print(m.span()) #span returns the position where the pattern group was found, returns a tuple of range
else:
  print("not found")
```

```python
#match always searches from the start of the string

s = "Something is there somewhere"

import re

m = re.search("s.*e",s,re.I) #search finds out only the first occurrance
if m != None: #m != None means that the pattern exists in the string
  print("Found")
  print(m.group()) #returns the group where the pattern was found
  print(m.span()) #span returns the position where the pattern group was found, returns a tuple of range
else:
  print("not found")

m = re.match("t.*e",s,re.I) #search finds out only the first occurrance
if m != None: #m != None means that the pattern exists in the string
  print("Found")
  print(m.group()) #returns the group where the pattern was found
  print(m.span()) #span returns the position where the pattern group was found, returns a tuple of range
else:
  print("not found")
```

```python
#to find all occurrences of the given pattern
s = "Somethis is there somewhere"
import re
#returns a list of portions of the string where pattern matches
lst = re.findall("s.*?e",s,re.I)
print(lst)

lst1 = re.finditer("s.*?e",s,re.I)
print(lst1) #callable_iterator object, can't be printed
if lst1 !=None:
  for m in lst1:
    print(m.group())
    print(m.span())
else:
  print('Not Found')
```

```python
import re

#it helps to create regex object to store regular expressions and flags in it
myreg = re.compile("s.*?e",re.I|re.M) #so that pattern and flags if provided will not have to be repeated
s = 'somethis is somewhere'
m = myreg.search(s)

if m != None:
  print(m.group())
  print(m.span())
else:
  print('not found')

news = re.sub('s.*?e','any',s,count=2,flags=re.I|re.M)
news = myreg.sub('any',s,count=2) #substitutes 'any' with the given pattern for the first two occurrances in the string

```
