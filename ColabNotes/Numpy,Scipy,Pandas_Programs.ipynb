*Pandas Series:
1. One-dimensional
2. Homogenous – Series elements must be of the same data type.
3. Size-immutable – Once created, the size of a Series object cannot be changed.

*Pandas DataFrame:
1. Two-dimensional
2. Heterogenous – DataFrame elements can have different data types.
3. Size-mutable – Elements can be dropped or added in an existing DataFrame.

```python
# We found null values at order id 3, hence we have to remove or replace these rows. This is called "Data Cleaning" or "Data Munging"
df['item_price1']=df['item_price'].map(lambda x:x.replace('$','0')) #to replace $ with 0
```

```python
df['item_price2']=df['item_price1'].astype('float')  #to convert from object data type to int
df['discounted_price']=df['item_price2']*0.85  #add new column
```

```python
df.pop('choice_description') # to delete the column
```

```python
index_nm=df[df['discounted_price']>3].index
df.drop(index_nm,inplace=True)  #drop all rows with discounted_price > 3 and ovewrite the original frame
df.drop([2,3,4],axis=0,inplace=True)
```
