**Dictionary**
- When you need to store the data in the key:value format then use dictionary

1. dictionary is represented as {}.
2. All the keys in the dictionary are unique and immutable objects, so a key can be of a type number,string,tuple, frozen set.
3. Dictionary is an ordered collection.
4. keys are unique but values can be different
5. keys can be used as index, so random access is possible with keys

*d["aa"]=30
- It will add a key value pair in dictionary if key is not there and if key is there then it will overwrite the value of the key.

*d.update(d1):
- It will update all the keys of d1 into dictionary d.
- d={"a":10,"b":100}
- d1={"b":200,"c":345,"d1":456}
- d.update(d1)
- then dictionary is d={"a":10,"b":200,"c":345,"d1":456}

** is expansion operator
d3={** d,**d1}:

*d.pop(key,[default value]):
- It will delete the key:value pair and return the value.But if key does not exists, and default value is not given then it will throw exception, but if key does not exists and default value is given, then it will given default value.

*d.popitem():
- It will delete the last key:value pair, and it will return a tuple (key:value)

*d.get(key[default value]):
- It will return the key value of the given key.But if key exists. but if the key is not there and default value isn't given, then will return None, and if key does not exist, and the default value is given then it will return default value.

*d.setdefault(key[default value]):
- if key exists, then it will return the value of the key but if the key doesn't key then it will add key:defaultvalue pair in the dictionary and returns default value.

*d.keys():
- It will return the list of all keys.

*d.values():
- It will return the list of all values.

*d.item():
- It will return iterable objects in which (key,value) are returned one by one, in the form of tuple.

*d.copy():
- It will create a shallow copy of the dictionary.

*d.clear():
- It will delete all (key,value) and keep empty dictionary.

*dict.fromkeys(lst,value):
- It will add all values from the list as keys in the dictionary and every key's value= data
- lst=(1,2,3)
- d=dict.fromkeys(lst,100) -> {1:100, 2:100, 3:100}

##Menu driven program using dictionary

```python
course={"DAI":(60,300),"DUASP":(40,350),"DAC":(240,400)}

# Case 1: Add new course
def addNewCourse():
    cname=input("Enetr Course Name: ")
    capacity=int(input("Enetr Capacity: "))
    duration=int(input("Enter Duration: "))
    v=course.get(cname,-1) # 	Optional. A value to return if the specified key does not exist. Default value None
    #if key is not there add it
    if v==-1:
        course[cname]=capacity,duration
        return True
    else:
        return False


# Case 2: Delete course
def deleteByName(cname):
    if cname in course:
        ans=input(f"{cname}, {course[cname]} Do you want to delete (y/n)? ")
        if ans=="y":
            course.pop(cname)
            return 1
        else:
            return 2
    else:
        return 3


# Case 3: Update course
def updateCourse(cname,capacity,duration):
    if cname in course:
        ans=input(f"{cname}, {course[cname]} Do you want to update(y/n)? ")
        if ans=="y":
            course[cname]=capacity,duration
            return 1
        else:
            return 2
    else:
        return 3


# Case 4: Display all the values from the given dictionary
def displayAll():
    for k,v in course.items():
        print(f"{k}---->{v}")


# Case 5: Display by capacity (capacity >= given capacity)
def findByCapacity(c):
    lst=[]
    for k,v in course.items():
        if v[0]>=c:
            lst.append((k,v))
    if len(lst)>0:
        return lst
    else:
        return None


# Case 6: Display data in sorted order of keys
def displayInSortedOrder():
    for k in sorted(course.keys()):
        print(f"{k} ---> {course[k]}")


# Case 7: Display by course name
def displayByName(cname):
    if cname in course:
        return course[cname]
    else:
        return None


# Case 8: Display by course duration
def displayByDuration(cduration):
    lst1=[]
    for k,v in course.items():
        if v[1]==cduration:
            lst1.append((k,v))
    if len(lst1)>0:
        return lst1
    else:
        return None


choice=0
while choice!=9:
    choice=int(input("""
    1. Add new course
    2. Delete course
    3. Update course
    4. Dispaly all
    5. Display by capacity
    6. Display in sorted order by name
    7. Display by course name
    8. Display by duration
    9. Exit
    Enter Choice: """))
    match choice:
        case 1:
            addNewCourse()

        case 2:
            cname=input("Enter course name to delete: ")
            status=deleteByName(cname)
            if status==1:
                print("Course found and deleted")
            elif status==2:
                print("Course found but not deleted")
            else:
                print("Course not found")

        case 3:
            cname=input("Enter course name to modify: ")
            capacity=int(input("Enetr capacity: "))
            duration=int(input("Enetr duration: "))
            status=updateCourse(cname,capacity,duration)
            if status==1:
                print("Course found and updated")
            elif status==2:
                print("Course found but not updated")
            else:
                print("Course not found")

        case 4:
            displayAll()

        case 5:
            c=int(input("Enter capacity to search: "))
            lst=findByCapacity(c)
            if lst!=None:
                for data in lst:
                    print(f"{data}")
            else:
                print("Data not found")

        case 6:
            displayInSortedOrder()

        case 7:
            cname=input("Enetr course name: ")
            c=displayByName(cname)
            if c!=None:
                print(f"{cname}--->{c}")
            else:
                print("Course not found")

        case 8:
            cduration=int(input("Enter capacity to search: "))
            lst1=displayByDuration(cduration)
            if lst1!=None:
                for data in lst1:
                    print(f"{data}")
            else:
                print("Data not found")

        case 9:
            print("Thank you for visiting...")

        case _:
            print("Invalid choice. Enter the right choice")
            pass
```
