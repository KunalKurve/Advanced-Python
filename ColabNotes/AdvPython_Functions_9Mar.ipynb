B. String Functions (Slicing in String)
*   When we use string in python, it is treated as list of characters.
*   THIS IS STRING is 0 to 13 or -14 to -1
*   str="THIS IS STRING"
*   str[-1]=G
*   str[5]or[-9]=I
*   str[3:9]=S IS S (to find charecters from 3rd to 8th position, exclude 9th)
*   str[2:]=IS IS STRING (Start from 2 to end)
*   str[:7]=THIS IS (Start from 0 to 6)
*   str[2:11:2]=I SSR (Start with 2nd index everytime increase position by 2, upto 10th index)
*   str[-11:10]=S IS ST (Reverese according to position)
*   str[::-1]=GNIRTS SI SIHT (Reverse the string)
*   STR[0::-1]= T
*   STR[0:-1:-1]

````python
str="THIS IS STRING"
print(str[0::-1])
````

````python
str="THIS IS STRING"
print(str[::-1])
print(str[-11:10])
print(str[:6])
print(str[3:10])
````

Functions in String
*   s.upper(): Convert string to uppercase.
*   s.lower(): Convert string to lowercase.
*   s.startswith("xxxx"): Return true if the string starts with xxxx otherwise false.
*   s.endswith("xxxx"): Return true if the string ends with xxxx otherwise false.
*   s.find(substr[start,end]): The index position of the first occurrence of the given substring if found, otherwise -1.
*   s.rfind(substr[start,end]): The index position of the last occurrence of the given substring if found, otherwise -1.
*   s.index(substr[start,end]): The index position of the first occurrence of the given substring if found, otherwise it returns exception.
*   s.rindex(substr[start,end]): The index position of the last occurrence of the given substring if found, otherwise it returns exception.
*   s.strip(set of characters): It will remove all occurrences of charecters from both sides of the string.
*   s.lstrip(set of characters): It will remove all occurrences of charecters from left side of the string.
*   s.rstrip(set of characters): It will remove all occurrences of charecters from right side of the string.
*   len(s): To find length of string
*   s.isdigit(): It will check all the characters in the string are digits.
*   s.isdecimal(): It will check all the characters in the string are decimal.
*   s.isnumber(): It will check all the characters in the string are numbers.
*   Ls.isalpha(): It will check all the characters in the string are alphabets.
*   s.isalnum(): It will check all the characters in the string are either alphabets or numbers.
*   s.count("abc"): Counts the number of occurrences of the string.
*   s.replace("abc","xyz"): Replace the occarrances everytime.
*   s.replace("abc","xyz",1): Replace the occarrance only once.
*   new_list = ":".join(lst): Joins each element of lst with ":" between them.

````python
s="She sells sea shells at the sea shore. sea is blue"
pos=s.find('sea')
print("Position=",pos)
pos=s.find('sea',pos+1)
print("Position=",pos)
pos=s.find('sea',pos+2)
print("Position=",pos)
pos=s.find('sea',pos+3)
print("Position=",pos)
````
