```python
from os import pidfd_open
class person:
  #init method is the constructor of any class
  #self is like this pointer in c++
  def __init__(self,pid=0,pname='',mob=''):
    self.name = pid
    # self.__pid = pid
    self.__pname = pname
    self.__mob = mob


  def __str__(self): #to display the whole object
    return f"ID: {self.name}, name: {self.__pname}, mobile: {self.__mob} "

  def set__pid(self,pid):
    self.__pid=pid
  def get__pid(self):
    return self.__pid
  def set__pname(self,pnm):
    self.__pname=pnm
  def get__pname(self):
    return self.__pname
  def set__mob(self,m):
    self.__mob=m
  def get__mob(self):
    return self.__mob

p1 = person(10,'Rajan','11111')
print(p1)
p1.set__mob('123123')
print(p1.get__mob())
# p1 = new person(10,'Rajan','11111') Wrong Syntax
# print(p1)
```
