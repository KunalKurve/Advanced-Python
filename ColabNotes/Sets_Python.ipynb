**Sets in Python**

- Sets are mutable. and collection of heterogeneous data.
- It stores only unique data.
- We use {} dictionary also, hence s={} will not create empty set. Created using the keyword 'set'. s=set()
- They are unordered.
- Case sensitive.
- Time complexity of searching is 1. (Hashing technique)
- Generally used where lot of searching is required and there are no duplicate values.
- Eg: Used for registration, where data input is needed only once.
- Indexing is not possible. Hence we have to search the value sequentially.
- There are no functions such as search, find, etc in set.
-
- Set maintains an internal hash table. (Usually mod function is used as we can store the result as the bucket number in the set. %4 gives values from 0 to 3)
- In sets it is easy to find duplicate values as it uses hashing technique.
- Stores only immutable data. It does not allow to store mutable data.
- We can store
- List, set and dictionary can not be stored in set. Because we will have to change hash table is that is true, which is not possible.
- s=set{}
- Using sets we can remove duplicate chars from the string.


```python
s="this is string"
s1=set(s)
print(s1)
```

**Functions in Set**

*len(set):
- Returns the length of the set.
*s.add(value):
- add value in the set if it does not exist. It can be added anywhere in the set.

*s.update(iterable):
- To add multiple values in the set.

*s.pop():
- It will detele the random value, and return that value.
- This will be last value according to the set.

*s.remove():
- To delete by value, then pass the value to remove().
- It will remove the value if exists, otherwise through an exception.

*s.discard():
- To delete by value, then pass the value to discard().
- It will remove the value if exists, otherwise it will ignore.

*s.copy():
- To create shallow copy of the set.

*clear():
- To delete all the values from the set.

**ABOVE VERSION 3**

*s.union(s1) or s|s1:
- Return a new set which has all the values from both the sets.

*s.intersection(s1) or s&s1:
- It will return the new set which has only the common values in both the sets.

*s.difference(s1) or s-s1:
- It returns a new set which has the values in s but not in s1.

*s.symmetric_difference(s1) or s^s1:
- It returns a new set which has all the values either only in s or only in s1.

*s.difference_update(s1):
- It will overwrite set s. It will replace it with all values which are only in s.

*s.symmetric_difference_update(s1) or s=s^s1:
- It will overwrite set s, which has all the values either only in s or only in s1.

*s.issubset(s1) or s<s1:
- If s1 has all the values of s and some extra values, then it will return true

*s.isdisjoint(s1):
- If no elements are common in sets then it will return true.

*s.issuperset(s1) or s>s1:
- If s1 has all the values of s and some extra values, then it will return true otherwise return false.

*s=frozenset(s):
- This will make the set immutable. Nothing can be changed inside a frozen set.
- This is used to secure the program.
