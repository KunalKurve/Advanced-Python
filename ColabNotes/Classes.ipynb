```python
class Person:
  #Constructor
  def __init__(self, pid, pname, mob):
    self.__pid = pid
    self.__pname = pname
    self.__mob = mob

  def __str__(self):
    return f"Id:{self.__pid} name:{self.__pname} mobile:{self.__mob}"

 #Getter and Setter methods
  def set_pid(self,pid):
    self.__pid=pid
  def get_pid(self):
    return self.__pid

  def set_pname(self,pname):
    self.__pname=pname
  def get_pname(self):
    return self.__pname

  def set_mob(self,mob):
    self.__mob=mob
  def get_mob(self):
    return self.__mob

p1 = Person(10,"Chetan","1111")
print(p1)
p1.set_mob("1234")
print(p1.get_mob())
```
