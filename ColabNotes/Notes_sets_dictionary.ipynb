```python
# prompt: write a code to display all the functions of sets in python and explain with comments

# Set functions in Python
s = {1, 2, 3, 4}

# add() function adds an element to the set.
s.add(5)
print(s)  # {1, 2, 3, 4, 5}

# update() function adds multiple elements to the set.
s.update([6, 7, 8])
print(s)  # {1, 2, 3, 4, 5, 6, 7, 8}

# remove() function removes an element from the set. returns None
s.remove(5)
print(s)  # {1, 2, 3, 4, 6, 7, 8}

# discard() function removes an element from the set if it exists.
s.discard(9)
print(s)  # {1, 2, 3, 4, 6, 7, 8}

# pop() function removes and returns an arbitrary element from the set.
s.pop()
print(s)  # {2, 3, 4, 6, 7, 8}

# clear() function removes all elements from the set.
s.clear()
print(s)  # set()

# union() function returns a new set with all elements from both sets.
s1 = {1, 2, 3}
s2 = {3, 4, 5}
s3 = s1.union(s2) # other syntax: s1|s2
print(s3)  # {1, 2, 3, 4, 5}

# intersection() function returns a new set with elements that are common to both sets.
s4 = s1.intersection(s2) # other syntax: s1&s1
print(s4)  # {3}

# difference() function returns a new set with elements that are in the first set but not in the second set.
s5 = s1.difference(s2) #s1-s2
print(s5)  # {1, 2}

# difference_update() function returns a new set with elements that are in the first set but not in the second set.
s5 = s1.difference_update(s2) #s1-s2
print(s5)  # {1, 2}

# symmetric_difference() function returns a new set with elements that are in either set but not in both sets. returns a new set
s6 = s1.symmetric_difference(s2) #s1^s2
print(s6)  # {1, 2, 4, 5}

# symmetric_difference_update() function overwrites set s which has all the values wither only in s1 or only in s2. does not return a new set
s6 = s1.symmetric_difference_update(s2) #s6 = s1^s2
print(s6)  # {1, 2, 4, 5}

# issubset() function returns True if the first set is a subset of the second set.
s7 = {1, 2}
print(s7.issubset(s1))  # True  #s7<s1

# issuperset() function returns True if the first set is a superset of the second set.
print(s1.issuperset(s7))  # True #s1>s7

# isdisjoint() function returns True if the two sets have no elements in common.
s8 = {5, 6, 7}
print(s1.isdisjoint(s8))  # True

#sets can be used in real life wherein something common has to be found out, when you only need unique elements,
# Eg: to find out the ATM fraud happening at ATMs at different locations, just by taking the intersection of the data of the transactions we can do it easily
```
