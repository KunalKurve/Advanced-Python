```python
lst=[] # empty list
lst=[10,20,30]
print("Length: ",len(lst))
#To retrieve the value at given index position
print(lst[2])
```

```python
lst1=[1,2,3,"xx"]
lst2=[12,13,"x",[100,200,300]]
print("To get the value from nested list: ",lst2[3][1])
#To overwrite the value
lst2[1]=100
print(lst2)
#To add values at end
lst2.append("sss")
print(lst2)
lst2.append([10,20])
print(lst2)

#To add multiple values in the lis at the end

lst1.extend("xyz")
print(lst1)

lst1.extend([10,20,30])
print(lst1)

#To add in between by position

lst1.insert(1,100)
print(lst1)
```

```python
lst3 = [1,2,3,"abc",[10,20,30],4,5]
#To add in a nested list
lst3[4].append(40)
print(lst3)

#To add multiple values at end in a nested list
lst3[4].extend([50,60])
print(lst3)

#To insert value in a nested list
lst3[4].insert(2,100)
print(lst3)
```
