**Regular_Expression**
- Site: regex101
- import re
- They are used for pattern matching.
- To design regex we use:

. :
- It matches with one character.

[a-z A-Z]:
- It matches with any alphabet.

[0-9] or \d:
- It matches with single digit.

*\w:
 - It matches with any word character [a-zA-Z0-9_].

\W:
 - It matches with any one character other than alphabets, digits and underscore [a-zA-Z0-9_].

\D:
 - Any nondigit char.

\b:
 - Any one boundry character (space).

\B:
 - Any one non boundry character.

\s:
- It matches with space.

\S:
- It matches with space.

*:
 - 0 or more occurences.

+:
 - 1 or more occurences.

?:
 - 0 or 1 occurence of preceding pattern.

{n}:
- exactly n occurances.

{m,n}:
- minimum m max n.

{m,}:
- min m.

^:
- It matches the pattern at the begining of the string.

$:
- It matches the pattern at the end of the string.



S = "Something is there somewhere."

[Ss].*e       ---> Something is there somewhere

[sS].*?e      ---> Some

\w+\s\w+\s\w+ ---> Something is there somewhere

^\w+\s\w+\s\w+$ ---> (The string should have exactly three words because of '^' and dollar) This is line (It won't accept any other string)

**Functions in Regex**

- It will find the first occurence of the given pattern,
2. match(pattern,string,flags):
3. findall(pattern,string,flags):
4. finditer(pattern,string,flags):
5. compile():
6. sub():



```python
import re
s = "Something is there somewhere"
m= re.search("s.*e",s,re.I)
if m!= None:
  print("Found")
  print(m.group())
  print(m.span())
else:
  print("Not found")
```

```python
# to find all occurences of the given pattern
s = "Something is there somewhere"

# Findall returns a list of string, where the portions of the string where pattern matches
lst= re.findall("s.*?e",s,re.I)
if lst != None:
  print(lst)
else:
  print("Not found")

print()

# Finditer is used to find position
lst1= re.finditer("s.*?e",s,re.I)
if lst1 != None:
  for m in lst1:
    print(m.group())
    print(m.span())
else:
  print("Not found")
```

```python
import re
s="This is home"
m=re.search("^(\w+)\s(\w+)\s(\w)+$",s,re.I)   #String with exactly 3 words
if m != None:
  print(m.group(1))
  print(m.span(1))
  print(m.group(2))
  print(m.span(2))
  print(m.group(3))
  print(m.span(3))
  print(m.group())
  print(m.span())

else:
  print("Not Found")
```
