**LIST**

- duplicate values allowed
- [ ]
- mutable and iterable (add, delete, modify, read) CRUD operations
- stores heterogeneous data
-lst= [10,12,13,5,7]
-lst= [10,12"xxx","yyy"]
-lst= [1,2,3,[100,200,300]]-> (list of 4)-> (lst[3][1]=200)
- we can perform all the perations on the inner list as that on the main list

*Append Function:
- add one value at the end of the list
- lst.append([100,200,300]) - creates a nested list

*Extend function:
- adds multiple values at a time at the end of the list
- lst.extend([100,200,300]) - makes 3 new elements

*Insert Function (pos , value)
- to insert a value in between  or anywhere in the list

*Pop Function;
- to delete the last value if position is not given, and a particular position if value is given.

*Remove Function
- to delete 1st occurance of an element from the list
- eg: lst2.remove(100)

*Count Function:
- to find number of occurances of the given value

*Clear Function:
- to remove all the elements from the list. keeps the empty list

*Index Function:
- to find the position of a given element

*Pos function:
- the value at the given position will be overwritten by the data

*Reverse Function:
- Reverses the given list


*Sort Function:
- In the list if all the values are of same type, then only sort is possible.
- It will sort in ascending order.
- For ascending-> lst.sort()
- For descending-> lst.sort(reverse=True)
- only possible on homogeneous data type
- if some value are float and some intgers it will work because they are numbers

*Copy Function:
- Creates a shallow copy of the list. It copies only references.
- lst1 = lst ; is a deep copy of the list lst ie lst1 points to the same memory location where lst is saved.

there is a copy module which has a deep copy function.

- Allows to store multiple values.
- it is immutable and iterable
- ()
- allows to store duplicate values

*Count Function:
- to find number of occurances of the given value

*Index Function:
- to find the position of a given element

*Length
- gives length of tuple.

*Uses
- To return multiple value from a function
- To pass variable number of parameters to a function
- Allows to store duplicate values
- To send variable number of parameters to a function
- A function can return multiple values as a tuple
